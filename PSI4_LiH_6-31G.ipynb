### Psi4 UHF 

```python
import psi4
psi4.core.be_quiet()
```

```python
psi4.set_memory('50000 MB')


molecule_geometry="""
Li 0.0 0.0 0.0
H {0} 0.0 0.0
symmetry c1
"""
            

opt={'scf_type': 'pk',
     'basis': '6-31g',
     'reference' :{},
     'freeze_core': True,
     'guess': 'huckel',
     'guess_mix' : True
                 }
```

```python
R1=[0.3,0.4,0.5,0.6,0.7,0.8,0.9,1.0,1.2,1.4,1.6,1.8,2.0,2.2,2.4,2.6,2.8,3.0,4.0,5.0]
R2=[1.5,1.51,1.52,1.53,1.54,1.55,1.56,1.57,1.58,1.59,1.6,1.61,1.62,1.63,1.64,1.65,1.66,1.67,1.68,1.69,1.7]
```

```python
opt['reference']='rhf'
fci=[]
rhf=[]
nuclear_repulsion_energy=[]
for i in range(len(R1)):
    r=R1[i]
    h2 = psi4.geometry(molecule_geometry.format(r))
    nuclear_repulsion_energy.append(h2.nuclear_repulsion_energy())
    mole_opt=psi4.set_options(opt)
    rhf.append(psi4.energy('scf'))
    fci.append(psi4.energy('fci'))
```

```python
opt['reference']='uhf'
uhf=[]
for i in range(len(R1)):
    r=R1[i]
    h2 = psi4.geometry(molecule_geometry.format(r))
    mole_opt=psi4.set_options(opt)
    uhf.append(psi4.energy('scf'))
```

```python
import pandas as pd
import numpy as np

dic = {
    "H-H Length": R1, 
    "RHF": rhf,
    "UHF": uhf,
    "Repulsion" :nuclear_repulsion_energy ,
    "FCI": fci,
    "FCI-Repulsion":np.array(fci)-np.array(nuclear_repulsion_energy)
}
df = pd.DataFrame(dic)
df
```
